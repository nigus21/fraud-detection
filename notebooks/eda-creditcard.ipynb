```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

credit_df = pd.read_csv("../data/raw/creditcard.csv")
credit_df.head()
```

```python
credit_df["Class"].value_counts()
```

```python
credit_df["Class"].value_counts(normalize=True)
```

```python
sns.countplot(x="Class", data=credit_df)
plt.title("Credit Card Fraud Class Distribution")
plt.show()
```

```python
sns.histplot(
    credit_df[credit_df["Class"] == 0]["Amount"],
    bins=50,
    kde=True,
    label="Legitimate"
)

sns.histplot(
    credit_df[credit_df["Class"] == 1]["Amount"],
    bins=50,
    kde=True,
    color="red",
    label="Fraud"
)

plt.legend()
plt.title("Transaction Amount Distribution")
plt.show()
```
